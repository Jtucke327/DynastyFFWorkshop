```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

team = "Los Angeles Rams"

team_needs = pd.DataFrame({
    'Team': ['Los Angeles Rams'],
    'Primary Needs': ['QB, RB, WR']
})
team_needs
```

```python
url = "https://www.sports-reference.com/cfb/years/2024-passing.html"
passing_df = pd.read_html(url)[0]
passing_df = passing_df[passing_df['Player'] != 'Player']
passing_df = passing_df[['Player', 'Team', 'Cmp', 'Att', 'Yds', 'TD', 'Int']]
passing_df.head()
```

```python
url = "https://www.sports-reference.com/cfb/years/2024-rushing.html"

rushing_df = pd.read_html(url)[0]
rushing_df.columns = [' '.join(col).strip() for col in rushing_df.columns]
rushing_df = rushing_df[rushing_df['Unnamed: 1_level_0 Player'] != 'Player']

rushing_df = rushing_df[[
    'Unnamed: 1_level_0 Player',
    'Unnamed: 2_level_0 Team',
    'Rushing Att',
    'Rushing Yds',
    'Rushing TD',
    'Receiving Rec',
    'Receiving Yds',
    'Receiving TD'
]]
rushing_df = rushing_df.rename(columns={
    'Unnamed: 1_level_0 Player': 'Player',
    'Unnamed: 2_level_0 Team': 'Team'
})

rushing_df.head()
```

```python
url = "https://www.sports-reference.com/cfb/years/2024-receiving.html"

receiving_df = pd.read_html(url)[0]
receiving_df.columns = [' '.join(col).strip() for col in receiving_df.columns]
receiving_df = receiving_df[receiving_df['Unnamed: 1_level_0 Player'] != 'Player']

receiving_df = receiving_df[[
    'Unnamed: 1_level_0 Player',
    'Unnamed: 2_level_0 Team',
    'Rushing Att',
    'Rushing Yds',
    'Rushing TD',
    'Receiving Rec',
    'Receiving Yds',
    'Receiving TD'
]]
receiving_df = receiving_df.rename(columns={
    'Unnamed: 1_level_0 Player': 'Player',
    'Unnamed: 2_level_0 Team': 'Team'
})

receiving_df.head()
```

```python
passing_df['Player'] = passing_df['Player'].str.strip()
rushing_df['Player'] = rushing_df['Player'].str.strip()
receiving_df['Player'] = receiving_df['Player'].str.strip()

passing_df['Position'] = 'QB'
rushing_df['Position'] = 'RB'
receiving_df['Position'] = 'WR'

master_df = pd.concat([passing_df, rushing_df, receiving_df], ignore_index=True)
```

```python
master_df['DraftScore'] = 0.0
print(master_df['Position'].unique())

master_df.loc[master_df['Position'] == 'QB', 'DraftScore'] = (
    master_df['Yds'] * 0.5 + 
    master_df['TD'] * 2 -
    master_df['Int'] * 1
)

master_df.loc[master_df['Position'] == 'RB', 'DraftScore'] = (
    master_df['Rushing Yds'] * 0.6 + 
    master_df['Rushing TD'] * 2 +
    master_df['Receiving Yds'] * 0.3
)

master_df.loc[master_df['Position'] == 'WR', 'DraftScore'] = (
    master_df['Receiving Yds'] * 0.7 + 
    master_df['Receiving TD'] * 2
)

top_candidates = master_df.sort_values(by='DraftScore', ascending=False)
top_candidates[['Player', 'Position', 'DraftScore']].head(10)
```
